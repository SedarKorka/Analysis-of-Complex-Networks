```python
# import the modules
import urllib.request
import networkx as nx
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#Load the dataset
# Define the URL of the edge list file for the Facebook-Ego network
facebook_url = "https://raw.githubusercontent.com/wang422003/Complex-Networks_exercise/main/Datasets/Group3/Facebook-Ego/348.edges"
# Download the Facebook-Ego network file
urllib.request.urlretrieve(facebook_url, "facebook.edges")
# Load the Facebook-Ego network
facebook_network = nx.read_edgelist("facebook.edges", nodetype=int)

# Define the URL of the edge list file for the Twitter-Ego network
twitter_url = "https://raw.githubusercontent.com/wang422003/Complex-Networks_exercise/main/Datasets/Group3/Twitter-Ego/789071.edges"
# Download the Twitter-Ego network file
urllib.request.urlretrieve(twitter_url, "twitter.edges")
# Load the Twitter-Ego network
twitter_network = nx.read_edgelist("twitter.edges", nodetype=int, create_using=nx.DiGraph())
```

```python
##Question a
# Get the number of nodes and edges in the Facebook-Ego network
facebook_num_nodes = len(facebook_network.nodes())
facebook_num_edges = len(facebook_network.edges())

# Get the number of nodes and edges in the Twitter-Ego network
twitter_num_nodes = len(twitter_network.nodes())
twitter_num_edges = len(twitter_network.edges())

# Print the results for Facebook-Ego network
print("Facebook-Ego Network:")
print("Number of nodes: ", facebook_num_nodes)
print("Number of edges: ", facebook_num_edges)

# Print the results for Twitter-Ego network
print("Twitter-Ego Network:")
print("Number of nodes: ", twitter_num_nodes)
print("Number of edges: ", twitter_num_edges)
```

```python
##Question b 

# Get the maximum degree and average degree for Facebook-Ego
max_degree_facebook = max(dict(facebook_network.degree()).values())
avg_degree_facebook = sum(dict(facebook_network.degree()).values()) / len(facebook_network)

# Get the maximum degree and average degree for Twitter-Ego
max_degree_twitter = max(dict(twitter_network.degree()).values())
avg_degree_twitter = sum(dict(twitter_network.degree()).values()) / len(twitter_network)

# Print the results
print("Facebook-Ego Network:")
print("Maximum degree:", max_degree_facebook)
print("Average degree:", avg_degree_facebook)

print("\nTwitter-Ego Network:")
print("Maximum degree:", max_degree_twitter)
print("Average degree:", avg_degree_twitter)
```

```python
##Question c
# Tracé du graphique
plt.figure(figsize=(10, 10))  # Réglage de la taille du graphique
pos = nx.spring_layout(facebook_network)  # Calcul des positions des nœuds
nx.draw(facebook_network, pos, with_labels=False, node_size=20, node_color='b', edge_color='gray', alpha=0.6)
plt.title("Graphique du réseau Facebook-Ego")
plt.show()

# Randomly select 5-8 nodes from Facebook-Ego
selected_nodes = random.sample(facebook_network.nodes(), random.randint(5, 8))

# Create a subgraph (partial network) using the selected nodes
partial_network = facebook_network.subgraph(selected_nodes)
# Tracé du sous-graphique
plt.figure(figsize=(8, 8))  # Réglage de la taille du graphique
pos = nx.spring_layout(partial_network)  # Calcul des positions des nœuds
nx.draw(partial_network, pos, with_labels=True, node_size=100, node_color='b', edge_color='gray', alpha=0.6)
plt.title("Sous-graphique du réseau Facebook-Ego")
plt.show()

# Calculate the adjacency matrix of the partial network using NumPy
adjacency_matrix = nx.to_numpy_array(partial_network)

# Print the adjacency matrix
print("Adjacency Matrix of Partial Network:")
print(adjacency_matrix)

# Explanation of the Need for an Adjacency Matrix:
print("\nExplanation:")
print("An adjacency matrix is a fundamental representation of a network that serves several important purposes:")
print("- Clarity of Structure: The adjacency matrix clearly shows which nodes are directly connected to each other and allows for a quick assessment of the network's connectivity patterns.")
print("- Efficient Computation: It simplifies various network analysis tasks, such as computing network metrics like centrality, shortest paths, and eigenvalues, using matrix operations.")
print("- Graph Algorithms: Many graph algorithms are naturally expressed in terms of matrix operations, making it easier to implement and analyze these algorithms.")
print("- Compact Representation: In cases where the network is sparse (i.e., has relatively few edges compared to the total possible connections), the adjacency matrix is a space-efficient way to represent the network.")

# Submit the adjacency matrix and explanation as part of the assignment.

print("We need the adjacency matrix to describe the structure of the network because it provides a compact and efficient representation of the network's connectivity. The adjacency matrix is a square matrix where the rows and columns correspond to the nodes in the network, and the entries indicate whether there is an edge between each pair of nodes. By examining the adjacency matrix, we can easily determine the degree of each node, the number of triangles in the network, and other important structural properties.")
```
